```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings as war
```

```python
df = pd.read_csv("/Users/satwikbiswas/Documents/Python/Data/Travel.csv")
df.head()
```

```python
df["Gender"].value_counts()
df["Gender"] = df["Gender"].replace("Fe Male","Female")
```

```python
df["MaritalStatus"].value_counts()
df["MaritalStatus"] = df["MaritalStatus"].replace("Single","Unmarried")
df["MaritalStatus"].value_counts()
```

```python
df["Designation"].value_counts()
```

```python
## finding features in side the df that have nul values
features_with_na = []
for feature in df.columns:
    if(df[feature].isnull().sum()>0):
        features_with_na.append(feature)
print(features_with_na)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["TypeofContact"] = df["TypeofContact"].fillna(df["TypeofContact"].mode()[0])
df["DurationOfPitch"] = df["DurationOfPitch"].fillna(df["DurationOfPitch"].median())
df["NumberOfFollowups"] = df["NumberOfFollowups"].fillna(df["NumberOfFollowups"].mode()[0])
df["PreferredPropertyStar"] = df["PreferredPropertyStar"].fillna(df["PreferredPropertyStar"].mode()[0])
df["NumberOfTrips"] = df["NumberOfTrips"].fillna(df["NumberOfTrips"].median())
df["NumberOfChildrenVisiting"] = df["NumberOfChildrenVisiting"].fillna(df["NumberOfChildrenVisiting"].mode()[0])
df["MonthlyIncome"] = df["MonthlyIncome"].fillna(df["MonthlyIncome"].median())
```

```python
df["NumberOfTotalVisiting"] = df["NumberOfPersonVisiting"]+df["NumberOfChildrenVisiting"]
df.drop(["NumberOfChildrenVisiting","NumberOfPersonVisiting"],inplace=True,axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(["ProdTaken"],axis=1)
y = df["ProdTaken"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=14,shuffle=True)

print(X_train.shape)
print(X_test.shape)
```

```python
## seperating the numerical and catagorical data

numerical_features = [feature for feature in df.columns if df[feature].dtype != "object"]
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'object']
print(numerical_features)
print(categorical_features)

numerical_features.remove("ProdTaken")
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

colTrans =  ColumnTransformer([
    ("cat",OneHotEncoder(handle_unknown="ignore",drop="first"),categorical_features),
    ("num",StandardScaler(),numerical_features)
])

X_train = colTrans.fit_transform(X_train)
X_test =  colTrans.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
models = {
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Logistic Regression": LogisticRegression(),
    "Adaboost": AdaBoostClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Corrected print statements
    print(f"Accuracy for {name} on train data: {accuracy_score(y_train, y_train_pred)*100}")
    print(f"Accuracy for {name} on test data: {accuracy_score(y_test, y_test_pred)*100}\n")
    print(classification_report(y_true=y_test,y_pred=y_test_pred))
    print("#"*100)
```

```python
rf_params = {
    "n_estimators": [50, 100, 200, 300, 400, 500],   # More options for tree count
    "max_depth": [10, 20, 30, 40, 50, None],        # Deeper trees for complex patterns
    "max_features": ["sqrt", "log2", None],         # Feature selection per split
    "bootstrap": [True, False],                     # Whether to use bootstrap sampling
    "min_samples_split": [2, 5, 10],                # Controls when to split a node
    "min_samples_leaf": [1, 2, 4]                   # Minimum samples in a leaf
}
ada_params = {
    "n_estimators": [50, 100, 200, 300],
    "learning_rate": [0.01, 0.1, 0.5, 1.0],
    "estimator": [DecisionTreeClassifier(max_depth=1), DecisionTreeClassifier(max_depth=3)]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_cv = [
    ["rf",RandomForestClassifier(n_jobs=-1),rf_params],
    ["ada",AdaBoostClassifier(),ada_params]
]

for name,model,param_grid in random_cv:

    random = RandomizedSearchCV(estimator=model,cv=5,random_state=12,n_jobs=-1,param_distributions=param_grid,scoring="accuracy")
    random.fit(X_train,y_train)
    print(f"{name}:")
    print(random.best_params_)
    print(random.best_score_)
    print(random.best_estimator_)
    y_pred = random.predict(X_test)
    print(classification_report(y_true=y_test,y_pred=y_pred))
    print("#"*100)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split


```

```python
diabetes = load_diabetes()
X = pd.DataFrame(diabetes["data"])
y = diabetes["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=10,shuffle=True,test_size=0.20)
print(X_train.shape,X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from sklearn.metrics import r2_score

models = {
    "Random Forest":RandomForestRegressor(n_jobs=-1,random_state=42, verbose=1),
    "Adaboost":AdaBoostRegressor(random_state=12)
}

for name,model in models.items():
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Corrected print statements
    print(f"Accuracy for {name} on train data: {r2_score(y_train, y_train_pred)*100}")
    print(f"Accuracy for {name} on test data: {r2_score(y_test, y_test_pred)*100}\n")
    print("#"*100)
```

```python
from sklearn.model_selection import GridSearchCV


rf_params = {
    "n_estimators": [50, 100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

rf_grid = GridSearchCV(RandomForestRegressor(n_jobs=-1, random_state=42), rf_params, cv=5,scoring="r2")
rf_grid.fit(X_train, y_train)

print("Best parameters for Random Forest:", rf_grid.best_params_)
print("BEst accuracy:",rf_grid.best_score_)
```

### Applying polynomial regression to check which is better for classification

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV


poly = PolynomialFeatures(include_bias=True,degree=2)
regression = LinearRegression(n_jobs=-1)


pipeline = Pipeline([
    ('poly-features', poly),
    ('line-reg', regression)
])


param_grid = {
    "poly-features__degree": list(range(1, 7))  # Degrees from 1 to 12
}


model = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, n_jobs=-1, scoring="r2")
model.fit(X_train, y_train)


print("Best Polynomial Degree:", model.best_params_["poly-features__degree"])
print("Best R² Score:", model.best_score_)
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

# Create a pipeline: Scaling + SVR
pipeline = Pipeline([
    ("scaler", StandardScaler()),  # Scaling is required for SVR
    ("svr", SVR())
])

# Define hyperparameter grid
param_grid = {
    "svr__C": [0.1, 1, 10, 100],  # Regularization parameter
    "svr__epsilon": [0.01, 0.1, 0.5, 1],  # Margin of error
    "svr__gamma": ["scale", "auto", 0.01, 0.1, 1]  # Kernel coefficient
}

# Perform Grid Search
model = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, n_jobs=-1, scoring="r2")
model.fit(X_train, y_train)

# Get best parameters and score
print("Best Parameters for SVR:", model.best_params_)
print("Best R² Score:", model.best_score_)

# Evaluate on test set
y_test_pred = model.best_estimator_.predict(X_test)
print("Test R² Score:", r2_score(y_test, y_test_pred))
```
