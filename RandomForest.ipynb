```python
## finding features in side the df that have nul values
features_with_na = []
for feature in df.columns:
    if(df[feature].isnull().sum()>0):
        features_with_na.append(feature)
print(features_with_na)
```

```python
## now we can fill the empty value of the numerical contents with mean or median and for catagorical value we can use mode
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["TypeofContact"] = df["TypeofContact"].fillna(df["TypeofContact"].mode()[0])
df["DurationOfPitch"] = df["DurationOfPitch"].fillna(df["DurationOfPitch"].median())
df["NumberOfFollowups"] = df["NumberOfFollowups"].fillna(df["NumberOfFollowups"].mode()[0])
df["PreferredPropertyStar"] = df["PreferredPropertyStar"].fillna(df["PreferredPropertyStar"].mode()[0])
df["NumberOfTrips"] = df["NumberOfTrips"].fillna(df["NumberOfTrips"].median())
df["NumberOfChildrenVisiting"] = df["NumberOfChildrenVisiting"].fillna(df["NumberOfChildrenVisiting"].mode()[0])
df["MonthlyIncome"] = df["MonthlyIncome"].fillna(df["MonthlyIncome"].median())
```

```python
df["NumberOfTotalVisiting"] = df["NumberOfPersonVisiting"]+df["NumberOfChildrenVisiting"]
df.drop(["NumberOfChildrenVisiting","NumberOfPersonVisiting"],inplace=True,axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(["ProdTaken"],axis=1)
y = df["ProdTaken"]
```

```python
df.isnull().sum()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=13,shuffle=True)

print(X_train.shape)
print(X_test.shape)
```

```python
## seperating the numerical and catagorical data

numerical_features = [feature for feature in df.columns if df[feature].dtype != "object"]
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'object']
print(numerical_features)
print(categorical_features)

numerical_features.remove("ProdTaken")
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

colTransfer = ColumnTransformer([
    ("cat",OneHotEncoder(handle_unknown="ignore",drop="first"),categorical_features),
    ("num",StandardScaler(),numerical_features)
])

X_train = colTransfer.fit_transform(X_train)
X_test = colTransfer.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

### Random Forest Classifier Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python

# Define models
models = {
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Logistic Regression": LogisticRegression()
}

# Train and evaluate models
for name, model in models.items():
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Corrected print statements
    print(f"Accuracy for {name} on train data: {accuracy_score(y_train, y_train_pred)*100}")
    print(f"Accuracy for {name} on test data: {accuracy_score(y_test, y_test_pred)*100}\n")
```

```python
param_grid = {
    "n_estimators": [50, 100, 200, 300, 400, 500],   # More options for tree count
    "max_depth": [10, 20, 30, 40, 50, None],        # Deeper trees for complex patterns
    "max_features": ["sqrt", "log2", None],         # Feature selection per split
    "bootstrap": [True, False],                     # Whether to use bootstrap sampling
    "min_samples_split": [2, 5, 10],                # Controls when to split a node
    "min_samples_leaf": [1, 2, 4]                   # Minimum samples in a leaf
}

random_cv=[
    ["RF",RandomForestClassifier(n_jobs=-1),param_grid]
]

from sklearn.model_selection import RandomizedSearchCV

for name,model,param_grid in random_cv:

    grid = RandomizedSearchCV(estimator=model,param_distributions=param_grid,cv=5,scoring="accuracy",n_jobs=-1)

    grid.fit(X_train,y_train)

    print(grid.best_params_)
    print(grid.best_score_)
    print(grid.best_estimator_)
    y_pred = grid.predict(X_test)
    print(classification_report(y_true=y_test,y_pred=y_pred))
```
