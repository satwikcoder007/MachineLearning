```python
df["Value"] = df["Value"].fillna(df["Value"].mean())
df
```

```python
df["Value"] = df["Value"]+2
df
```

```python
df.groupby("Product")["Value"].mean() ## this will group the products and will print the mean of each group
```

```python
df.groupby(["Product","Region"])["Value"].sum() ## this will basically select all the unique product and give their region wise value sum
```
